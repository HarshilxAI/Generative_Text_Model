```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Load the tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")

# Set model to evaluation mode
model.eval()
```

```python
print(generate_text("The role of Artificial Intelligence in healthcare"))
```

```python
def generate_text(prompt, max_length=150):
    inputs = tokenizer.encode(prompt, return_tensors="pt")

    # Generate text with pad_token_id set
    outputs = model.generate(
        inputs,
        max_length=max_length,
        num_return_sequences=1,
        do_sample=True,
        temperature=0.7,
        top_k=50,
        top_p=0.95,
        pad_token_id=tokenizer.eos_token_id  # 🔧 Fix warning
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
print(generate_text("How climate change is affecting our daily lives"))
```

```python
# Ask the user for a topic
user_prompt = input("Enter a topic or starting sentence: ")

# Generate the paragraph
output = generate_text(user_prompt)

# Show the result
print("\n📝 Generated Paragraph:\n")
print(output)
```

```python
print(generate_text("Time management tips for college students"))
```
